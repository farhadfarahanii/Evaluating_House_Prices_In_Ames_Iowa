```python
# Function for plotting histogram
def his_dis(column, titlename, x, bins, color):    
    plt.figure(figsize=(10,6))
    plt.hist(column, bins=bins, edgecolor="black", color=color)
    plt.title("Distribution of " + titlename)
    plt.xlabel(x)
    plt.ylabel("Frequencies");
```

```python
# Function for plotting number of values in each column
def plot_values(column, title, color):    
    plt.figure(figsize=(7,4))
    column.value_counts().plot(kind="barh", color=color, edgecolor="black")
    plt.title(f"The Number of Each Value in {title}", fontsize=12)
    plt.xlabel("Number of Values", fontsize=10)
    plt.ylabel("Values", fontsize=10);
```

## Null Values
The task is to identify the null values in each column of a dataset and make a decision on what to do with them. When working with datasets, it's common to encounter columns that contain null or missing values. One approach to handling these missing values is to drop the entire column if it has a high percentage of null values.

To determine if a column should be dropped, we typically set a threshold for the percentage of null values that we are willing to tolerate. In this case, the threshold is set to 10%. This means that if a column has more than 10% null values, we will consider dropping it.

Above plots give us insights about the type of the values in each columns:
- For columns Garage Qual, Garage Cond, Bsmt Cond, and Bsmt Qual, we will replace null values with "TA" (average) since it represents the majority value and is a reasonable imputation technique.

Outliers are data points that are significantly different from other data points in a dataset. These can occur due to errors in data collection or measurement, or due to the presence of unusual or extreme values in the population being studied.

One way to identify outliers is to use scatter plots, which represents the relationship between two variables. Scatter plots can help to identify patterns and trends in the data, and can also highlight any data points that fall outside of the expected range.

Once we have identified outliers, we will remove them from the dataset to help to improve the accuracy of our analysis and modeling by reducing the impact of extreme values.

## Categorical Variables
When working with categorical variables, it can be useful to first plot the frequency of each value in the variable. This can help us to identify any values that occur very rarely or that are not representative of the broader population.

If a categorical variable has very little diversity, meaning that most of the values fall into one or two categories, it may not be useful for our analysis. In this case, we may choose to drop the variable entirely, as it is unlikely to provide meaningful information.

However, in some cases, a categorical variable may have many different possible values, but some of these values may occur very rarely. These low-frequency values may not provide enough information to be useful on their own, but they may still contain some valuable information when combined with other similar values.

To address this issue, we can create a new category called "Other" and group all of the low-frequency values into this category. This can help to simplify the dataset and reduce the number of categories we need to consider in our analysis.
